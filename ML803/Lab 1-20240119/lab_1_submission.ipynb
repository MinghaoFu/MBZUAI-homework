```python
# task 2
import numpy as np
import matplotlib.pyplot as plt
P1_mean_list = []

for i in range(1000):
    P1 = np.random.normal(0, 1, 4)

    # Calculate the new sample mean
    P1_mean = np.mean(P1)
    P1_mean_list.append(P1_mean)  
plt.hist(P1_mean_list, bins=100)
std = np.std(P1_mean_list)
print('empirical standard deviation: ', std)
      
```
