```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr

# Constants
N = 1000
a = 0.5

# Part (a) - Both X and E are standard Gaussian variables
# Generate data
np.random.seed(1)  # for reproducibility
X = np.random.normal(0, 1, N)
E = np.random.normal(0, 1, N)
Y = a * X + E

# i) Scatter plot of (X, E) and (X, Y)
plt.figure(figsize=(12, 6))

# Scatter plot of X and E
plt.subplot(1, 2, 1)
plt.scatter(X, E, alpha=0.5)
plt.title('Scatter plot of X and E')
plt.xlabel('X')
plt.ylabel('E')

# Scatter plot of X and Y
plt.subplot(1, 2, 2)
plt.scatter(X, Y, alpha=0.5)
plt.title('Scatter plot of X and Y')
plt.xlabel('X')
plt.ylabel('Y')

plt.tight_layout()
plt.show()
```

```python
# ii) Perform linear regression of Y on X
lr_y_on_x = LinearRegression()
lr_y_on_x.fit(X.reshape(-1, 1), Y)

# Coefficient estimate
beta_x = lr_y_on_x.coef_[0]

# Calculate the estimated noise term
estimated_noise_y_on_x = Y - lr_y_on_x.predict(X.reshape(-1, 1))

# Scatter plot of X and the estimated noise term
plt.figure(figsize=(6, 4))
plt.scatter(X, estimated_noise_y_on_x, alpha=0.5)
plt.title('Scatter plot of X and estimated noise term (Y on X)')
plt.xlabel('X')
plt.ylabel('Estimated Noise')
plt.axhline(y=0, color='r', linestyle='--')  # Highlight zero line
plt.show()


# Test for independence using Pearson correlation
correlation, p_value = pearsonr(X, estimated_noise_y_on_x)

print("Correlation Coefficient:", correlation)
print("P-value:", p_value)

beta_x
```

```python
# iii) Regress X on Y
lr_x_on_y = LinearRegression()
lr_x_on_y.fit(Y.reshape(-1, 1), X)

# Coefficient estimate
beta_y = lr_x_on_y.coef_[0]

# Calculate the estimated noise term
estimated_noise_x_on_y = X - lr_x_on_y.predict(Y.reshape(-1, 1))

# Scatter plot of Y and the estimated noise term
plt.figure(figsize=(6, 4))
plt.scatter(Y, estimated_noise_x_on_y, alpha=0.5)
plt.title('Scatter plot of Y and estimated noise term (X on Y)')
plt.xlabel('Y')
plt.ylabel('Estimated Noise')
plt.axhline(y=0, color='r', linestyle='--')  # Highlight zero line
plt.show()

correlation, p_value = pearsonr(Y, estimated_noise_x_on_y)

print("Correlation Coefficient:", correlation)
print("P-value:", p_value)

beta_y
```
