```python
# task 6
import numpy as np
import scipy.stats as stats

def bootstrap_median_ci(sample, alpha=0.05, num_bootstraps=1000):
    bootstrapped_medians = []
    n = len(sample)
    for _ in range(num_bootstraps):
        bootstrap_sample = np.random.choice(sample, size=n, replace=True)
        bootstrapped_medians.append(np.median(bootstrap_sample))
    lower = np.percentile(bootstrapped_medians, (alpha/2)*100)
    upper = np.percentile(bootstrapped_medians, (1-alpha/2)*100)
    return lower, upper

# Parameters
num_samples = 1000
sample_size = 30
alpha = 0.05

mean_coverage = 0
median_coverage = 0
mean_widths = 0
median_widths = 0

for _ in range(num_samples):
    sample = np.random.normal(0, 1, sample_size)

    # Confidence interval for mean
    se_mean = np.std(sample, ddof=1) / np.sqrt(sample_size)
    ci_mean = stats.norm.interval(1-alpha, loc=np.mean(sample), scale=se_mean)
    mean_widths += ci_mean[1] - ci_mean[0]
    if ci_mean[0] <= 0 <= ci_mean[1]:
        mean_coverage += 1

    # Confidence interval for median (bootstrap)
    ci_median = bootstrap_median_ci(sample, alpha=alpha)
    median_widths += ci_median[1] - ci_median[0]
    if ci_median[0] <= 0 <= ci_median[1]:
        median_coverage += 1

mean_coverage /= num_samples
median_coverage /= num_samples
mean_widths /= num_samples
median_widths /= num_samples

mean_coverage, median_coverage, mean_widths, median_widths
```

```python
# task 7
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def normal_ci(sample, alpha=0.05):
    p_hat = np.mean(sample)
    se = np.sqrt(p_hat * (1 - p_hat) / len(sample))
    ci = stats.norm.interval(1 - alpha, loc=p_hat, scale=se)
    return ci

def hoeffding_ci(sample, alpha=0.05):
    p_hat = np.mean(sample)
    n = len(sample)
    h = np.sqrt(np.log(2/alpha) / (2*n))
    ci = (max(p_hat - h, 0), min(p_hat + h, 1))
    return ci

sample_sizes = [20, 100, 1000]
p_values = [0.2, 0.4, 0.5, 0.8]
alpha = 0.05

plt.figure(figsize=(15, 10))
plot_number = 1

for p in p_values:
    for n in sample_sizes:
        sample = np.random.binomial(1, p, n)

        ci_normal = normal_ci(sample, alpha=alpha)
        ci_hoeffding = hoeffding_ci(sample, alpha=alpha)

        plt.subplot(len(p_values), len(sample_sizes), plot_number)
        plt.errorbar(1, np.mean(sample), yerr=[[np.mean(sample)-ci_normal[0]], [ci_normal[1]-np.mean(sample)]], fmt='o', label='Normal CI')
        plt.errorbar(2, np.mean(sample), yerr=[[np.mean(sample)-ci_hoeffding[0]], [ci_hoeffding[1]-np.mean(sample)]], fmt='o', label='Hoeffding CI')
        plt.xticks([1, 2], ['Normal', 'Hoeffding'])
        plt.title(f'n={n}, p={p}')
        plt.ylim(0, 1)
        if plot_number == 1:
            plt.legend()

        plot_number += 1

plt.tight_layout()
plt.show()
```
