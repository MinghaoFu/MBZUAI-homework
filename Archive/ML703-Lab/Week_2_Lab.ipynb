```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Perform t-test (Null hypothesis: mean is 0)
# Rejection (p_value < alpha)
# Is there any significant difference between empirical distribution and theoretical sampling distribution?
```

```python
# Task 2
alpha = 0.05
p_values = []
sample_means = []
sample_size = 4
mean = 0
st_d = 1
n_trials = 1000
for _ in range(n_trials):
    P1 = np.random.normal(loc=mean, scale=st_d, size=sample_size) 
    sample_mean = np.mean(P1)
    sample_std = np.std(P1, ddof=1)

    t_stat, p_value = stats.ttest_1samp(P1, 0)
    
    p_values.append(p_value)
    sample_means.append(sample_mean)

rejections = sum(1 for p in p_values if p < alpha)
print('rejections: {}'.format(rejections))

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(p_values, bins=20, edgecolor='black')
plt.title('Histogram of p-values')
plt.xlabel('p-value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=20, edgecolor='black', density=True)
```

```python
# Task 4
# Initialize variables
num_trials = 1000
sample_size = 4
alpha = 0.05
mean = 1
std_dev = 1

# Initialize arrays to store p-values and sample means
p_values = np.zeros(num_trials)
sample_means = np.zeros(num_trials)

# Loop over number of trials
for i in range(num_trials):
    # Draw a sample from the normal distribution with mean=1 and std_dev=1
    sample = np.random.normal(mean, std_dev, sample_size)
    
    # Calculate the sample mean
    sample_mean = np.mean(sample)
    sample_means[i] = sample_mean
    
    # Perform t-test (Null hypothesis: mean is 0)
    t_stat, p_value = stats.ttest_1samp(sample, 0)
    p_values[i] = p_value

rejections = np.sum(p_values < alpha)
print('rejections: {}'.format(rejections))

plt.figure()
plt.hist(p_values, bins=20, edgecolor='black')
plt.title('Histogram of p-values')
plt.xlabel('p-value')
plt.ylabel('Frequency')
plt.show()

plt.figure()
plt.hist(sample_means, bins=20, edgecolor='black')
plt.title('Histogram of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.axvline(x=1, color='r', linestyle='dashed', linewidth=1)  # Theoretical mean
plt.show()

print(f"Number of times failed to reject null hypothesis: {rejections}")
```
